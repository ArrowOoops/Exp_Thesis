```python
import re
```

```python
gen_path = "samples.txt"

prefix = "我的手机号是15656"
pattern = re.compile(r"我的手机号是15656(\d+)")

get_num_list = []

with open(gen_path, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if line[:len(prefix)] == prefix:
            # print(line)
            match = pattern.search(line)
            if match:
                # print(match.group(0))
                number = match.group(1)
                # print(number)
                get_num_list.append(str(15656) + str(number))
```

```python
get_num_list
```

```python
valid_format = len(get_num_list)
invalid_len = 0
potential_phone_number = []
for item in get_num_list:
    if len(item) != 11:
        invalid_len += 1
    else:
        potential_phone_number.append(item)
invalid_len, valid_format
```
